```python
# To run this program, dataset in article version is necessary

!pip install contractions

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
import contractions
from collections import defaultdict

nltk.download('punkt')
nltk.download('stopwords')
CLASS_LEVEL = 0
```

```python
# You can use this class to load articles and recieve feedback in testing
class Game:
    def __init__(self, class_value):
        self.words = defaultdict(int)
        self.class_val = class_value

    def tokenize(self, article):
        # data cleaning, lemmerization...
        # Change the following code with the data cleaning and preprocessing code you used
        article = article.lower()
        article = article.replace('\n', ' ')
        article = article.replace('=', '')
        article = article.replace('[^a-zA-Z\']', ' ')
        article = contractions.fix(article)
        article = article.replace(r'\s+', ' ')
        tokens = word_tokenize(article)
        stop_words = set(stopwords.words('english'))
        # Change the code above with the data cleaning and preprocessing code you used

        for token in tokens:
            if token not in stop_words and token not in ['.', '(', ')', ',']:
                # lemmatizer or any other pre-process
                self.words[token] += 1

    def guessWord(self, word):
        if word in self.words:
            return self.words[word]
        return 0

    def guessClass(self, val):
        return val == self.class_val

    def test_print(self):
        print(self.words)
        print(len(self.words))
        print(self.class_val)
```

```python
# This is a test part for Game class
dataset = pd.read_csv('/content/drive/MyDrive/544/Group Dataset/Final Dataset/articles/dataset.csv')
question = dataset.sample(n=1)
class_file = open('/content/drive/MyDrive/544/Group Dataset/Final Dataset/articles/classes.txt', 'r')
class_list = class_file.readlines()[CLASS_LEVEL].strip('\n').split('|')
game_simulator = Game(class_list[question.iloc[0, CLASS_LEVEL]])
game_simulator.tokenize(question.iloc[0, -1])
game_simulator.test_print()
```
